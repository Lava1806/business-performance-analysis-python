```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x= 'Category', y= 'Sales', data=df)
plt.title("Sales by Category")
plt.show()
```

```python
category_analysis = df.groupby('Category')[['Sales','Profit']].sum()
category_analysis
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x= 'Discount', y= 'Profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

```python
Insight: Higher discount levels negatively impact profitability, indicating that aggressive discounting reduces margins. 
```

```python
region_analysis = df.groupby('Region')[['Sales','Profit']].sum()
region_analysis
```

```python
Insight: Sales performance varies across regions, with some regions contributing significantly more revenue than others.
```

```python
sns.barplot(x='Region', y='Sales', data=df)
plt.title("Sales by Region")
plt.show()
```

```python
pivot = df.pivot_table(values='Profit', index='Category', columns='Region', aggfunc='sum')

sns.heatmap(pivot, annot=True, cmap='coolwarm')
plt.title("Profitability Heatmap")
plt.show()
```

```python
## KEY INSIGHTS

- High discounting reduces profitability
- Certain categories have high sales but low profit
- Regional performance is uneven
- Revenue is concentrated among key customers
```

```python
## BUSINESS RECOMMENDATIONS

- Optimize discount strategies to protect margins
- Focus on high-margin product categories
- Improve performance in underperforming regions
- Strengthen relationships with high-value customers
```
